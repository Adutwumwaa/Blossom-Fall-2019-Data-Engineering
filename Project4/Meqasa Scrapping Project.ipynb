```python
import requests
```

```python
pages = (range(28))
```

```python
results = []
for p in pages:
    outcome = requests.get('https://meqasa.com/houses-for-rent-in-Accra?w={}'.format(p+1))
    results.append(outcome)
```

```python
from bs4 import BeautifulSoup
```

```python
total_soup = [] 
for result in results:
    soup = BeautifulSoup(result.content, 'lxml')
    total_soup.append(soup)
    
    
```

```python
total_listings = []
for soup in total_soup:
    listings = soup.find_all("div", {"class":"mqs-prop-dt-wrapper"})
    total_listings.append(listings)
```

#Extracting info from page 1 of 28 - Total_listings_1

```python
total_listings_1 = total_listings[0]
```

```python
def escape(a):
    if a == None:
        return 0 
    else:
        return a.text
    

```

```python

base_url = 'https://meqasa.com' 
listings_1 =[]
for listings in total_listings_1:
    title = listings.find('h2').text.replace('\n', '')
    listings_1.append(title)
    bed = escape(listings.find('li',{'class':'bed'}))
    listings_1.append(bed)
    shower = escape(listings.find('li',{'class':'shower'}))
    listings_1.append(shower)
    garage = escape(listings.find('li',{'class':'garage'}))
    listings_1.append(garage)
    area = escape(listings.find('li',{'class':'area'}))
    listings_1.append(area)
    description = escape(listings.find_all('p')[1])
    listings_1.append(description)
    x = listings.find_all('p')[0].text.replace('\n', '').replace(':', '').replace('PriceGH₵', 'GH₵ ').replace('Price$','$ ').strip('').split(' ')
    currency = x[0]
    listings_1.append(currency)
    price = x[1]
    listings_1.append(price)
    rent_period = x[2]
    listings_1.append(rent_period)
    url = base_url + listings.find('a').attrs['href'] 
    listings_1.append(url)
    address = listings.find('h2').text.replace('\n', '').split("at")[1]
    listings_1.append(address)
    time_posted = listings.find('p',{'class':'wsnr'})
    listings_1.append(time_posted)
```

```python
#this breaks the resulting list into subsets of 12 items each,
def group(l, size):
     return [l[i:i+size] for i in range(0, len(l), size)]
```

```python
glistings_1 = group(listings_1, 12)
```
